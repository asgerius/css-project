```python
import pandas as pd
import datetime
import vaderSentiment as vs
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

```python
data = pd.read_pickle(r'data.pkl')
data['creation_date'] = pd.to_datetime(data['creation_date'],unit="s")
```

```python
analyzer = SentimentIntensityAnalyzer()
sentiment=[]
for text in data['clean']:
    vs = analyzer.polarity_scores(text)
    sentiment.append(vs)

data['Sentiment'] = sentiment
```

```python
data = pd.concat([data.drop(['Sentiment'],axis=1), data['Sentiment'].apply(pd.Series)], axis=1)
```

```python
data.groupby('creation_date').mean().plot(kind='bar',x='creation_date', y='compound')
```
