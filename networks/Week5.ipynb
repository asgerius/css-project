```python
### Ex. 15: Write a program to find all words that occur at least three times in the Brown Corpus.
brown_vocab_count = dict((word, 0) for word in set(brown.words()))

for w in brown.words():
    brown_vocab_count[w] += 1
# print(brown_vocab_count)

above_three_words = []

for word, count in brown_vocab_count.items():
    if count >= 3:
        above_three_words.append(word)
# print(above_three_words)
print("There are {0} unique words in brown corpus.".format(len(set(brown.words()))))
print("{0} of these words appear three or more times in the corpus.".format(len(above_three_words)))
```

```python
### Ex. 17: Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

def fifty_most_freq_words(text):
    stopwords = nltk.corpus.stopwords.words()
    frequencies = nltk.FreqDist(w for w in text if w not in stopwords and w.isalpha())
    return frequencies.most_common(50)
#fifty_most_freq(text1).tabulate()
print(fifty_most_freq_words(text1))
```

```python
### Ex. 18: Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

def fifty_most_freq_bigrams(text):
    stopwords = nltk.corpus.stopwords.words()
    frequencies = nltk.FreqDist(b for b in nltk.bigrams(text) if b[0] not in stopwords and b[1] not in stopwords and b[0].isalpha() and b[1].isalpha())
    return frequencies.most_common(50)
#fifty_most_freq(text1).tabulate()
print(fifty_most_freq_bigrams(text1))
```
